```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Sinh dữ liệu hàm số
# Tạo 256 điểm trong khoảng từ -pi đến pi
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

# 2. Kiểm tra dữ liệu (In ra console)
print(f'+ X: |X|={len(X)}, x in [{np.min(X):.2f}, {np.max(X):.2f}]')
print(f'  10 mẫu đầu: {X[0:10]}')

# 3. Hiển thị đồ thị
plt.figure(figsize=(10, 6)) # Chỉnh kích thước đồ thị cho dễ nhìn

# Vẽ đồ thị Cos và Sin với nhãn (label) để làm chú thích
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

# Thiết lập giới hạn trục X và các điểm đánh dấu (ticks) bằng ký hiệu LaTeX
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Thiết lập giới hạn trục Y
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

# --- PHẦN THÊM MỚI ĐỂ ĐỒ THỊ ĐẸP HƠN ---
plt.title("Đồ thị hàm số Sin và Cos") # Tiêu đề
plt.legend(loc='upper left')          # Hiển thị chú thích (sine/cosine)
plt.grid(True, linestyle='--', alpha=0.6) # Thêm lưới mờ cho dễ quan sát
plt.axhline(0, color='black',linewidth=1) # Thêm trục hoành đen đậm
plt.axvline(0, color='black',linewidth=1) # Thêm trục tung đen đậm

plt.show()
```
