```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Tạo dãy số
print("1. Tạo dãy số a")
a = np.array([120, 183, 160, 175, 145, 162, 190, 160, 152, 162])
print("a =", a)

# 2. Các phép toán cơ bản
print("\n2. Tính toán trên dãy")

# Bình phương
a2 = a ** 2
print("Bình phương:", a2)

# Độ dài
print("Độ dài:", len(a))

# Max, Min, Trung bình
print("Max:", np.max(a))
print("Min:", np.min(a))
print("Trung bình:", np.mean(a))

# Phương sai & độ lệch chuẩn
mean = np.mean(a)
variance = np.sum((a - mean) ** 2) / (len(a) - 1)
std = np.sqrt(variance)
print("Phương sai:", variance)
print("Độ lệch chuẩn:", std)
```

```python
# 3. Tạo dãy x
x = np.arange(1, 11)
print("\nx =", x)

# Lọc số chẵn
a_chan = a[a % 2 == 0]
x_chan = x[x % 2 == 0]

print("a chẵn:", a_chan)
print("x chẵn:", x_chan)
```

```python
# Vị trí lẻ
idx = list(range(1, len(a), 2))

a_le = a[idx]
x_le = x[idx]

diff = a_le - x_le
distance = np.sqrt(np.sum(diff ** 2))

print("\nVị trí lẻ:", idx)
print("Hiệu:", diff)
print("Khoảng cách:", distance)
```

```python
min_dist = np.min(np.abs(a[:, np.newaxis] - x))
print("\nKhoảng cách nhỏ nhất:", min_dist)
```

```python
# Sinh ngẫu nhiên
x_rand = np.random.randint(1, 11, 10000)
value, count = np.unique(x_rand, return_counts=True)

print("\nGiá trị:", value)
print("Tần số:", count)

# Vẽ
plt.bar(value, count)
plt.title("Phân bố đều")
plt.show()
```

```python
m = 1
s = 1

x = np.random.rand(100) * 10 - 5
y = (1/(s*np.sqrt(2*np.pi))) * np.exp(-((x-m)**2)/(2*s**2))

plt.scatter(x, y)
plt.title("Phân phối chuẩn")
plt.show()
```

```python
x = m + np.random.randn(100000) * s

count, bins = np.histogram(x, bins=300, density=True)
bins = (bins[1:] + bins[:-1]) / 2

plt.scatter(bins, count)
plt.title("Histogram chuẩn")
plt.show()
```

```python
A = np.random.randint(1, 6, (4,3))
B = np.random.randint(1, 6, (4,3))
X = np.random.randint(1, 6, (3,4))

print("\nA:\n", A)
print("B:\n", B)
print("X:\n", X)

print("A+B:\n", A+B)
print("A-B:\n", A-B)
print("A*B:\n", A*B)

Y = np.matmul(A, X)
print("A.X:\n", Y)
```

```python
A = np.array([[4,3,-5], [-2,-4,5], [8,8,0]])
y = np.array([2,5,-3])

x = np.linalg.solve(A, y)

print("\nNghiệm x:", x)

# kiểm tra
print("Ax =", np.matmul(A, x))
```

```python
import math

def calc_pi(n):
    points = np.random.rand(n, 2) * 2 - 1
    inside = np.sum(points**2, axis=1) <= 1
    return 4 * np.sum(inside) / n

print("\nSai số:")
print("n=100:", calc_pi(100) - math.pi)
print("n=10000:", calc_pi(10000) - math.pi)
print("n=1000000:", calc_pi(1000000) - math.pi)
```
