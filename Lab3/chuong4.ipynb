```python
import csv
import numpy
import requests

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
response = requests.get(url)
lines = response.text.splitlines()
reader = csv.reader(lines, delimiter=',', quoting=csv.QUOTE_NONE)
x = list(reader)
data_standard = numpy.array(x).astype('float')
print("Standard Library - Kích thước:", data_standard.shape)
```

```python
from numpy import loadtxt
import urllib.request

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
raw_data = urllib.request.urlopen(url)
data_numpy = loadtxt(raw_data, delimiter=",")
print("NumPy - Kích thước:", data_numpy.shape)
```

```python
import pandas
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pandas.read_csv(url, names=names)
print("Pandas - Kích thước:", data.shape)
```
