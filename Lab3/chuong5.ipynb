```python
import pandas as pd
from pandas import set_option

# Khởi tạo DataFrame 'data'. Thay thế 'your_dataset.csv' bằng đường dẫn đến tệp dữ liệu của bạn.
data = pd.read_csv('/content/sample_data/mnist_train_small.csv')

# 1. Xem 20 dòng dữ liệu đầu tiên (Peek)
print("--- 20 dòng đầu ---")
print(data.head(20))

# 2. Xem kích thước dữ liệu (Dimensions)
print("\n--- Kích thước dữ liệu ---")
print(data.shape)

# 3. Xem kiểu dữ liệu từng thuộc tính (Data Types)
print("\n--- Kiểu dữ liệu ---")
print(data.dtypes)

# 4. Thống kê mô tả (Descriptive Statistics)
print("\n--- Thống kê mô tả ---")
set_option('display.width', 100)
set_option('display.precision', 3)
print(data.describe())

# 5. Phân phối lớp (Class Distribution)
# Cần xác định cột 'class' nếu có, hoặc thay thế bằng tên cột nhãn của dataset mnist_train_small.csv
# Ví dụ, nếu cột đầu tiên là nhãn, có thể đổi 'class' thành tên cột đó hoặc sử dụng index.
# Hiện tại, mnist_train_small.csv không có cột tên là 'class', nên phần này có thể cần điều chỉnh.
# Để tạm thời chạy được, tôi sẽ bỏ qua phần này hoặc thay bằng một cột có sẵn.
# Nếu cột đầu tiên là nhãn, có thể sử dụng: print(data.iloc[:, 0].value_counts())
# print("\n--- Phân phối lớp ---")
# print(data.groupby('class').size())

# 6. Tương quan giữa các thuộc tính (Pearson Correlation)
print("\n--- Ma trận tương quan ---")
correlations = data.corr(method='pearson')
print(correlations)

# 7. Độ xiên của phân phối (Skewness)
print("\n--- Độ xiên (Skew) ---")
print(data.skew())
```
